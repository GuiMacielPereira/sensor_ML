```python
# Notebook for processing raw data 
# NOTE: This notebook should be run after the script that converts raw data to npz format
# This notebook takes raw data in npz format and selects triggers 

import numpy as np
import matplotlib.pyplot as plt
from peratouch.config import data_dir
from peratouch.preprocessing import SelectTriggers

# Width of triggers 
width = 32    # No of samples

# Path of raw data
path = data_dir / "raw_npz" / "five_users.npz"
data = np.load((path))
```

```python
# Look at the signal collected for the first user
key = list(data.keys())[0] 
ST = SelectTriggers(data)
ST.plot_signal(key)
```

```python
# Run routine that finds indexes of triggers and selects only suitable triggers
ST.run()
```

```python
# Plot some examples of triggers selected
ST.plot_clean(key)
```

```python
# Plot some triggers (concatenated in a single plot) that are not included in the analysis 
# First plot is noisy triggers, second is short triggers
ST.plot_discarded(key)
```

```python
# Compare the mean and std of the signals for all the users involved
ST.plot_means_std()
```

```python
# Save file
filename = path.name.split(".")[0] + "_window" + f"_{ST.get_triggers()[key].shape[1]}.npz"
save_path = data_dir / "processed" / filename
ST.save_dict(save_path)
```
